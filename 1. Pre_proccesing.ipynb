```python
#all the imports
import pandas as pd
import json
```

```python
#read path and filenames from json
with open('private.json', 'r') as fp:
    data = json.load(fp)
    
#
path = data["path"]
original_csv = data["original-csv"]
clean_csv = data["clean-csv"]
```

```python
#read the original data
df = pd.read_csv(path + original_csv)
```

```python
#deep copy, so we will always have the original copy in hand
flies = df.copy()
```

```python
#create the column with the conditioning from the first dir
conditionings = ["Assa_Males_Grouped_and_Female", "Assa_Males_Grouped_Unknown_","Assa_Males_single","Assa_Males_Grouped_FirstDay", "Assa_Males_Grouped_SecondDay", "Assa_Males_Grouped_ThirdDay",  "Grouped_Starved", "Mated", "Rejected", "Single"]

def substitute(arg):
    
    for conditioning in conditionings:
        if conditioning.upper() in arg.upper():
            return conditioning

    return arg


flies["dir"] = flies["dir"].apply(substitute)
flies.rename(columns = {"dir" : "conditioning"}, inplace = True)
```

```python
flies['conditioning'].unique()
```

```python
flies['conditioning'].replace(['Assa_Males_Grouped_and_Female'], 'Males_Grouped_and_Female', inplace = True)
flies['conditioning'].replace(['Assa_Males_Grouped_FirstDay'], 'Grouped', inplace = True)
flies['conditioning'].replace(['Assa_Males_Grouped_SecondDay'], 'Grouped', inplace = True)
flies['conditioning'].replace(['Assa_Males_Grouped_ThirdDay'], 'Grouped', inplace = True)
flies['conditioning'].replace(['Assa_Males_Grouped_Unknown_'], 'Grouped', inplace = True)
flies['conditioning'].replace(['Assa_Males_single'], 'Single', inplace = True)
```

```python
flies['conditioning'].unique()
```

```python
flies.head()
```

```python
#change fly.?? column name into values.??
cols = flies.columns
for c, col in enumerate(cols):
    
    if "values" in col:
        name = flies.iloc[0, c - 2]
        p = name.find(".mat")
        if(p > 0):
            name = name[:p]
           
        flies.rename(columns = {col : name}, inplace = True)
```

```python
flies.head()
```

```python
#drop all "fly", "files", and "dir" columns
fly_columns = [c for c in flies.columns if "fly" in c[:3]]
flies.drop(fly_columns, axis = 1, inplace = True)

files_columns = [c for c in flies.columns if "file" in c[:4]]
flies.drop(files_columns, axis = 1, inplace = True)

dir_columns = [c for c in flies.columns if "dir" in c[:3]]
flies.drop(dir_columns, axis = 1, inplace = True)
```
