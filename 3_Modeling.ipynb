```python
#all the imports
import json
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.tree import export_graphviz
from math import sqrt
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
#read path and filenames from json
import json
with open('private.json', 'r') as fp:
    data = json.load(fp)
    
#
path = data["path"]
chosen_csv = data["chosen-csv"]
```

```python
#read the clean data
flies = pd.read_csv(path + chosen_csv, index_col = 0)
```

```python
#flies.shape
flies.columns
```

```python
flies_dummies = pd.get_dummies(flies, prefix = '', prefix_sep = '')
```

```python
flies_dummies.head()
```

```python
def model_scatter(model, X_train, X_test, y_train, y_test, caption, ax):
    y_train_pred = model.predict(X_train)
    y_test_pred  = model.predict(X_test)

    rmse_train = sqrt(mean_squared_error(y_train, y_train_pred))
    rmse_test  = sqrt(mean_squared_error(y_test, y_test_pred))

    r2_train = r2_score(y_train, y_train_pred)
    r2_test  = r2_score(y_test, y_test_pred)
    
    print(f"{caption}\n\trmse/r2 train =  {rmse_train:}/{r2_train}\n\trmse/r2 test =  {rmse_test}/{r2_test}")

    train = pd.DataFrame({"y" : y_train, "y_pred" : y_train_pred})
    train["label"] = f"train; rmse: {rmse_train:.3f}; r2: {r2_train:.3f}"

    test = pd.DataFrame({"y" : y_test, "y_pred" : y_test_pred})
    test["label"] = f"test;  rmse: {rmse_test:.3f}; r2: {r2_test:.3f}"
    
    df = pd.concat([train, test])
    sns.scatterplot(data = df, x = "y", y = "y_pred", hue = "label", ax = ax).set(title = caption)
    plt.setp(ax.get_legend().get_texts(), fontsize='22') # for legend text
    plt.setp(ax.get_legend().get_title(), fontsize='32') # for legend title
   
    plt.savefig("figs.pdf", bbox_inches='tight', transparent = True)
```

```python
def regression(score : str, trainsize : float= 0.7, dominant_count : int = 3, laso_alpha : float = 0.01, scatter_plot : bool = True):

    #start pipeline
    X = flies_dummies.drop([score], axis = 1)
    y = flies_dummies[score]

    #Splitting to Test/Train
    X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = trainsize, random_state = 0)

    #Testing with Lasso
    lasso = Lasso(alpha = laso_alpha).fit(X_train, y_train)
    abs_coef = np.abs(lasso.coef_)
    coef_index = np.argsort(abs_coef)
    print(f"Lasso |Coefs|:\n{abs_coef}\n{coef_index}")    
    
    #selecting only the dominant features
    keep_columns = set()
    for i in range(dominant_count):
        keep_columns.add(X_test.columns[coef_index[-i]])
    print(f"Selecting {dominant_count} Columns:\n{keep_columns}")
    drop_columns = set(X_test.columns.tolist()) - keep_columns
    #print(f"Dropping Columns:\n{drop_columns}")
    X_train = X_train.drop(drop_columns, axis = 1)
    X_test = X_test.drop(drop_columns, axis = 1)

    lr = LinearRegression().fit(X_train, y_train)    
    knr = KNeighborsRegressor(n_neighbors = dominant_count).fit(X_train, y_train)
    dtr = DecisionTreeRegressor().fit(X_train, y_train) #max_depth = dominant_count

    if scatter_plot:

        fig, ax = plt.subplots(3, 1, figsize = (20, 20))

        #Linear Regression Fitting
        model_scatter(lr, X_train, X_test, y_train, y_test, "Linear Regression", ax[0])        

        #KNN Fitting
        model_scatter(knr, X_train, X_test, y_train, y_test, "K Neighbors Regression", ax[1])

        #Descision tree
        model_scatter(dtr, X_train, X_test, y_train, y_test, "Decision Tree Regression", ax[2])
    
    return (lr, knr, dtr, list(keep_columns))
```

```python
regression("Grooming", trainsize = 0.7, dominant_count = 10);
```

```python
regression("Chase", trainsize = 0.8, dominant_count = 7);
```

```python
regression("Chain", trainsize = 0.9, dominant_count = 4)
```

```python
regression("Aggregation", trainsize = 0.9, dominant_count = 3)
```

```python
(lr, knr, dtr, features) = regression("Grooming", trainsize = 0.9, dominant_count = 3, scatter_plot = False)

#tree output that does not require graphviz
plt.figure(figsize = (15, 15))
tree.plot_tree(dtr, max_depth = 3, feature_names = features, filled = True, fontsize = 20);
```

## Decision Tree with optimal min_samples_leaf number based on loss function (msls vs RMSE_test)

```python
X = flies_dummies.drop('Chain', axis=1)
y = flies_dummies['Chain']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

chain_model = DecisionTreeRegressor(min_samples_leaf = 4).fit(X_train, y_train)


y_train_pred = chain_model.predict(X_train)
rmse_train = np.sqrt(mean_squared_error(y_train, y_train_pred))
print(f"Chain_train_ rmse = {rmse_train}")

y_test_pred = chain_model.predict(X_test)
rmse_test = np.sqrt(mean_squared_error(y_test, y_test_pred))
print(f"chain_test_rmse = {rmse_test}")

r2_train = r2_score(y_train, y_train_pred)
print(f"train r_2 = {r2_train}")
r2_test = r2_score(y_test, y_test_pred)
print(f"test r_2 = {r2_test}")

feature_important = []

for feature, importance in zip(X.columns, chain_model.feature_importances_):
  if importance > 0:
        feature_important.append(feature)

        print(f'{feature:12}: {importance}')
number_of_important = len(feature_important)
print(f'{number_of_important} important features: {feature_important}')

features1 = list(X.columns)
features1

train = pd.DataFrame({"y" : y_train, "y_pred" : y_train_pred})
train[""] = f"train; rmse: {rmse_train:.3f}; r2: {r2_train:.3f}"

test = pd.DataFrame({"y" : y_test, "y_pred" : y_test_pred})
test[""] = f"test;  rmse: {rmse_test:.3f}; r2: {r2_test:.3f}"
    
df = pd.concat([train, test])
fig = plt.figure()
ax = fig.gca()
sns.scatterplot(data = df, x = "y", y = "y_pred", hue = "", ax = ax).set(title = 'Chain')
plt.setp(ax.get_legend().get_texts(), fontsize='12') # for legend text
plt.setp(ax.get_legend().get_title(), fontsize='12') # for legend title


plt.savefig("chain tree.pdf", bbox_inches='tight', transparent = True)
```

```python
plt.figure(figsize = (15, 15))
tree.plot_tree(chain_model, max_depth = 3, feature_names = features1, filled = True, fontsize = 20);
```

Using Linear Regression to test the correlation between Aggregation behavior and Chain behavior

```python
from sklearn.model_selection import train_test_split as split
flies_dummies_train, flies_dummies_test = split(flies_dummies, test_size=0.33, random_state=42)

x_train = flies_dummies_train[['Aggregation']]
y_train = flies_dummies_train['Chain']

x_test = flies_dummies_test[['Aggregation']]
y_test = flies_dummies_test['Chain']

agg_model = LinearRegression().fit(x_train, y_train)

y_train_pred = agg_model.predict(x_train)
rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
print(f"agg-chain_ rmse = {rmse}")

y_test_pred = agg_model.predict(x_test)
rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))
print(f"agg-chain_rmse = {rmse}")

r2 = r2_score(y_train, y_train_pred)
print(f"train r_2 = {r2}")
r2 = r2_score(y_test, y_test_pred)
print(f"test r_2 = {r2}")

train = pd.DataFrame({"y" : y_train, "y_pred" : y_train_pred})
train["label"] = "train"  

test = pd.DataFrame({"y" : y_test, "y_pred" : y_test_pred})
test["label"] = "test" 

df = pd.concat([train, test])
plt.figure()
sns.scatterplot(data = df, x = "y", y = "y_pred", hue = "label").set(title = 'agg-chain')

plt.savefig("agg-chain.pdf", bbox_inches='tight', transparent = True)
```
